# $\mathbf{Z}2$ Lattice Gauge Theory

Data Analysis Notebook

The data are generated by running the `main.c` file, which calls the `z2_example.c` script. This script was originally written by [Michael Creute](https://quark.phy.bnl.gov/~creutz/), and has been modified. It outputs pairs of ($\beta$, $\overline{E}$), which we analyze here for continuity.

## Imports

```python
import numpy
import matplotlib.pyplot as plt
import pandas

# Configure LaTeX support in plotting
# plt.rc('text', usetex=False)
```

## Load Data

```python
data = pandas.read_csv('~/repos/research/CGFT/data.csv').values
```

## Configure Plot

```python
# Split data
mid_pt = data.shape[0] // 2
beta, energy = data[:-1,0], data[:-1,1]

# Setup plot
fig, ax = plt.subplots()
ax.plot(beta[:mid_pt], energy[:mid_pt], label='Heating', color='red')
ax.plot(beta[mid_pt:], energy[mid_pt:], label='Cooling', color='blue')
ax.set(xlabel=r'Beta ($\beta$)', 
       ylabel=r'Magnetization ($M$)',
       title='Z2 Magnetization vs Beta')
ax.grid()
fig.savefig("test.pdf")
plt.show()
```
